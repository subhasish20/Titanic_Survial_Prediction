```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Feature Engineering
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from keras.layers import Dense
```

```python
# taking the important column
data = df[[
    'pclass',
    'sex',
    'age',
    'sibsp',
    'parch',
    'fare',
    'embarked',
    'survived']]
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.age.mean()
```

```python
data.age.median()
```

```python
data['age'] = data['age'].fillna(data.age.mean())
```

```python
data.age.isnull().sum()
```

```python
data.isnull().sum()
```

```python
data.fare.median()
```

```python
data.fare.mean()
```

```python
data['fare'] = data.fare.fillna(data.fare.mean())
```

```python
data.isnull().sum()
```

```python
data.embarked.value_counts()
```

```python
data['embarked'] = data['embarked'].fillna('S')
```

```python
data.isnull().sum()
```

```python
data
```

```python
x = data.drop(['survived'],axis=1)
```

```python
y = data['survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train
```

```python
ct = ColumnTransformer([
    ("tnf1",StandardScaler(),['age','fare']),
    ("tnf2",OneHotEncoder(),['sex','embarked'])
],remainder='passthrough')

```

```python
"""feature_names = ct.get_feature_names_out()

x_train_scaled_df = pd.DataFrame(x_train_scaled, columns=feature_names)

print(x_train_scaled_df.head())"""
#x_train_scaled = ct.fit_transform(X_train)
#x_test_scaled = ct.transform(X_test)
```

```python
pipe_decission_tress = Pipeline(steps=[
    ('preprocessor', ct),
    ('classifier', DecisionTreeClassifier())
])
```

```python
pipe_decission_tress.fit(X_train,y_train)
```

```python
pred = pipe_decission_tress.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
pipeline_logistic = Pipeline(steps=[
    ('preprocessor', ct),
    ('classifier', LogisticRegression())
])
```

```python
pipeline_logistic.fit(X_train,y_train)
```

```python
log_pred = pipeline_logistic.predict(X_test)
```

```python
accuracy_logistic = accuracy_score(y_test,log_pred)
print(f'Logistic Regression Accuracy: {accuracy_logistic:.2f}')
```

```python
pipeline_random_forest = Pipeline(steps=[
    ('preprocessor', ct),
    ('classifier', RandomForestClassifier())
])
```

```python
pipeline_random_forest.fit(X_train,y_train)
```

```python
y_pred_rf = pipeline_random_forest.predict(X_test)
```

```python
y_pred_rf
```

```python
accuracy_score(y_test,y_pred_rf)
```

```python
pipeline_knn = Pipeline(steps=[
    ('preprocessor', ct),
    ('classifier', KNeighborsClassifier())
])
```

```python
pipeline_knn.fit(X_train,y_train)
```

```python
y_pred_knn = pipeline_knn.predict(X_test)
```

```python
y_pred_knn
```

```python
print(accuracy_score(y_pred_knn,y_test))
```
